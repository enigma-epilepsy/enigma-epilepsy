```python
# Initialize a geocoder
geolocator = Nominatim(user_agent="my_geocoding_app")
```

```python
# Create a world map using OpenStreetMap.Humanitarian tile layer with attribution
# m = folium.Map(location=[40.0, -100.0], zoom_start=4, tiles='OpenStreetMap',
#                attr='© OpenStreetMap contributors, Humanitarian OpenStreetMap Team')

#m = folium.Map(location=[40.0, -100.0], zoom_start=4, tiles='Esri.NatGeoWorldMap')
m = folium.Map(location=[30.0, 10.0], zoom_start=2, tiles='Esri.WorldImagery',
               attr="ENIGMA-Epilepsy Consortium - Esri/Google Map")

# Add markers for each city
for city in cities:
    location = geolocator.geocode(city)
    if location:
        latitude, longitude = location.latitude, location.longitude
        marker = folium.Marker([latitude, longitude], popup=city).add_to(m)

display(m)
```
